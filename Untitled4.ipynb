```python
def get_mw(features_labels):
    j = 1
    match_week = []
    for i in range(len(features_labels)):
        match_week.append(j)
        if ((i + 1) % 10) == 0:
            j = j + 1
    features_labels['MW'] = match_week
    return features_labels
```

```python
from copy import deepcopy

def get_match_result(is_home_team, FTR):
  if FTR == 'H':
    return 'W' if is_home_team else 'L'
  elif FTR == 'A':
    return 'L' if is_home_team else 'W'
  else:
    return 'D'

def get_teams_last_3_games_form(features_labels):
  teams = {key:['M'] for key in features_labels.HomeTeam.unique()} # 'M' since it is the begining of the season
  
  for i, row in features_labels.iterrows():
    
    # add accumlated home team goals based on previous matches
    teams[row.HomeTeam].append(get_match_result(is_home_team=True, FTR=row.FTR))
      
    # add accumlated away team goals based on previous matches
    teams[row.AwayTeam].append(get_match_result(is_home_team=False, FTR=row.FTR))
    
  
  teams2 = deepcopy(teams)
  teams3 = deepcopy(teams)
 
  for key in teams:
    teams2[key].insert(0,'M')
    
    teams3[key].insert(0,'M')
    teams3[key].insert(0,'M')
    
    
  # initialize last 3 games to 'M'
  for i in range(1, 3):
    features_labels['HM' + str(i)] = 'M'
    features_labels['AM' + str(i)] = 'M'
  
  for i, row in features_labels.iterrows():
    
    features_labels.at[i, 'HM3'] = teams3[row.HomeTeam].pop(0)
    features_labels.at[i, 'HM2'] = teams2[row.HomeTeam].pop(0)
    features_labels.at[i, 'HM1'] = teams[row.HomeTeam].pop(0)
    
    features_labels.at[i, 'AM3'] = teams3[row.AwayTeam].pop(0)
    features_labels.at[i, 'AM2'] = teams2[row.AwayTeam].pop(0)
    features_labels.at[i, 'AM1'] = teams[row.AwayTeam].pop(0)
  
  return features_labels
```

```python
def get_match_point(is_home_team, FTR):
  if FTR == 'H':
    return 3 if is_home_team else 0
  elif FTR == 'A':
    return 0 if is_home_team else 3
  else:
    return 1

def get_accumlated_points_by_team(features_labels):
  teams = {key:[0] for key in features_labels.HomeTeam.unique()}
  
  for i, row in features_labels.iterrows():
    
    # add accumlated home team goals based on previous matches
    teams[row.HomeTeam].append(teams[row.HomeTeam][-1] + get_match_point(is_home_team=True, FTR=row.FTR))
      
    # add accumlated away team goals based on previous matches
    teams[row.AwayTeam].append(teams[row.AwayTeam][-1] + get_match_point(is_home_team=False, FTR=row.FTR))
    
  features_labels['HTP'] = 0
  features_labels['ATP'] = 0
  for i, row in features_labels.iterrows():
    
    features_labels.at[i, 'HTP'] = teams[row.HomeTeam].pop(0)
    features_labels.at[i, 'ATP'] = teams[row.AwayTeam].pop(0)
  
  return features_labels

features_labels = get_accumlated_points_by_team(features_labels)
```

```python
def transformHM1(row):
    if(row.HM1 == 'W'):
        return 3
    elif(row.HM1 == 'L'):
        return 0
    else:
        return 1
def transformHM2(row):
    if(row.HM2 == 'W'):
        return 3
    elif(row.HM2 == 'L'):
        return 0
    else:
        return 1
def transformHM3(row):
    if(row.HM3 == 'W'):
        return 3
    elif(row.HM3 == 'L'):
        return 0
    else:
        return 1
def transformAM1(row):
    if(row.AM1 == 'W'):
        return 3
    elif(row.AM1 == 'L'):
        return 0
    else:
        return 1
def transformAM2(row):
    if(row.AM2 == 'W'):
        return 3
    elif(row.AM2 == 'L'):
        return 0
    else:
        return 1
def transformAM3(row):
    if(row.AM3 == 'W'):
        return 3
    elif(row.AM3 == 'L'):
        return 0
    else:
        return 1
    
    
    
```

```python
features_labels["HM1"] = features_labels.apply(lambda row: transformHM1(row),axis=1)
features_labels["HM2"] = features_labels.apply(lambda row: transformHM2(row),axis=1)
features_labels["HM3"] = features_labels.apply(lambda row: transformHM3(row),axis=1)
features_labels["AM1"] = features_labels.apply(lambda row: transformAM1(row),axis=1)
features_labels["AM2"] = features_labels.apply(lambda row: transformAM2(row),axis=1)
features_labels["AM3"] = features_labels.apply(lambda row: transformAM3(row),axis=1)
```

```python
def get_accumlated_goals_scored_by_team(features_labels):
  teams = {key:[0] for key in features_labels.HomeTeam.unique()}
  
  for i, row in features_labels.iterrows():
    
    # add accumlated home team goals based on previous matches
    teams[row.HomeTeam].append(teams[row.HomeTeam][-1] + row.FTHG)
      
    # add accumlated away team goals based on previous matches
    teams[row.AwayTeam].append(teams[row.AwayTeam][-1] + row.FTAG)
  
  
  features_labels['HTGS'] = 0
  features_labels['ATGS'] = 0
  for i, row in features_labels.iterrows():
    
    features_labels.at[i, 'HTGS'] = teams[row.HomeTeam].pop(0)
    features_labels.at[i, 'ATGS'] = teams[row.AwayTeam].pop(0)
  
  return features_labels
```

```python
def get_accumlated_goals_conceded_by_team(features_labels):
  teams = {key:[0] for key in features_labels.HomeTeam.unique()}
  
  for i, row in features_labels.iterrows():
    
    # add accumlated home team goals based on previous matches
    teams[row.HomeTeam].append(teams[row.HomeTeam][-1] + row.FTAG)
      
    # add accumlated away team goals based on previous matches
    teams[row.AwayTeam].append(teams[row.AwayTeam][-1] + row.FTHG)
  
  
  features_labels['HTGC'] = 0
  features_labels['ATGC'] = 0
  for i, row in features_labels.iterrows():
    
    features_labels.at[i, 'HTGC'] = teams[row.HomeTeam].pop(0)
    features_labels.at[i, 'ATGC'] = teams[row.AwayTeam].pop(0)
  
  return features_labels
```
