```python
import requests
import numpy as np
```

```python
standings_url = "https://fbref.com/en/comps/9/2021-2022/schedule/2021-2022-Premier-League-Scores-and-Fixtures"
```

```python
data = requests.get(standings_url)
```

```python
import pandas as pd
matches = pd.read_html(data.text)[0]
```

```python
matches[139:147]
```

```python
matches["Score"] = matches['Score'].apply(str)
matches["Score"] = matches['Score'].apply(list)
```

```python
matches.head(11)
matches = matches[matches.Wk.notnull()]
matches.head(11)
```

```python
matches.reset_index(inplace=True)
matches.reset_index(inplace=True)
matches
```

```python
df = pd.DataFrame(matches[['Score','level_0']])
df[['FTHG','vs','FTAG']] = pd.DataFrame(df.Score.tolist(), index= df.index)
df
```

```python
match = matches.drop('Score', axis=1)
match
```

```python
df['FTHG'] = df['FTHG'].replace('n', -1)
df['FTAG'] = df['FTAG'].replace('n', -1)
df
```

```python
df_1 = df.drop('Score', axis=1)
```

```python
df_2 = df_1.drop('vs', axis=1)
df_2.head(100)
```

```python
w = pd.merge(match, df_2, on ='level_0')
w.tail(200)
```

```python
epl = w[['level_0','Wk','Home','Away','FTHG','FTAG','xG','xG.1']]
```

```python
epl["FTHG"] = epl['FTHG'].apply(int)
epl["FTAG"] = epl['FTAG'].apply(int)
epl['xG'] = epl['xG'].replace(np.nan, -1)
epl['xG.1'] = epl['xG.1'].replace(np.nan, -1)
epl['xG'] = epl['xG'].apply(float)
epl['xG.1'] = epl['xG.1'].apply(float)
epl['xG'] = epl['xG'].replace(-1, 0)
epl['xG.1'] = epl['xG.1'].replace(-1, 0)
```

```python
epl['HomeTeam'] = epl['Home']
epl['AwayTeam'] = epl['Away']
feat_table = epl[['level_0','Wk','HomeTeam','AwayTeam','FTHG','FTAG','xG','xG.1']]
feat_table
```

```python
def Result(row):
    if(row.FTHG > row.FTAG):
        return 'H'
    elif(row.FTHG < row.FTAG):
        return 'A'
    elif(row.FTHG == -1):
        return np.nan
    else:
        return 'D'
feat_table["Result"] = feat_table.apply(lambda row: Result(row),axis=1)
feat_table
```

```python
stat = pd.read_csv("E0.21_22.csv")
stat['HomeTeam'] = stat['HomeTeam'].replace('Newcastle', 'Newcastle Utd').replace('Leeds', 'Leeds United').replace('Leicester','Leicester City').replace("Norwich","Norwich City").replace('Man United','Manchester Utd').replace('Man City','Manchester City')
stat['AwayTeam'] = stat['AwayTeam'].replace('Newcastle', 'Newcastle Utd').replace('Leeds', 'Leeds United').replace('Leicester','Leicester City').replace("Norwich","Norwich City").replace('Man United','Manchester Utd').replace('Man City','Manchester City')
stat.head(10)
```

```python
feat_table["FTHG"] = epl['FTHG'].replace(-1, 0)
feat_table["FTAG"] = epl['FTAG'].replace(-1, 0)
```

```python
feat_table["Match"] = feat_table['HomeTeam'] +"-"+ feat_table['AwayTeam']
stat["Match"] = stat['HomeTeam'] +"-"+ stat['AwayTeam']
```

```python
new_df = pd.merge(feat_table, stat, on='Match', how='left')
new_df[125:145]
    






    
    
```

```python
epl_stat=new_df[['Div','Date','Time','Wk','HomeTeam_x','AwayTeam_x','Result','FTHG_x','FTAG_x','xG','xG.1','HS','AS','HST','AST']]
```

```python
epl_stat[110:130]
```

```python
epl_stat['FTHG'] = epl_stat['FTHG_x']
epl_stat['FTAG'] = epl_stat['FTAG_x']
epl_stat['HxG'] = epl_stat['xG']
epl_stat['AxG'] = epl_stat['xG.1']
epl_stat['HomeTeam'] = epl_stat['HomeTeam_x']
epl_stat['AwayTeam'] = epl_stat['AwayTeam_x']
```

```python
epl_stat=epl_stat[['Wk','HomeTeam','AwayTeam','Result','FTHG','FTAG','HxG','AxG','HS','AS','HST','AST']]
```

```python
epl_stat.dtypes
```

```python
epl_stat["pastHG"] = 0.0
epl_stat["pastAG"] = 0.0
epl_stat["pastHxG"] = 0.0
epl_stat["pastAxG"] = 0.0
epl_stat["pastHS"] = 0.0
epl_stat["pastAS"] = 0.0
epl_stat["pastHST"] = 0.0
epl_stat["pastAST"] = 0.0
```

```python
k = 3
for i in range(0,366):
    row = epl_stat.loc[i]
    ht = row.HomeTeam
    at = row.AwayTeam
    ht_stats = epl_stat.loc[0:i-1][(epl_stat.HomeTeam == ht) | (epl_stat.AwayTeam == ht)].tail(k)
    at_stats = epl_stat.loc[0:i-1][(epl_stat.HomeTeam == at) | (epl_stat.AwayTeam == at)].tail(k)
    epl_stat.at[i, 'pastHG'] = (ht_stats[ht_stats["AwayTeam"] == ht].sum().FTAG + ht_stats[ht_stats["HomeTeam"] == ht].sum().FTHG)
    epl_stat.at[i, 'pastAG'] = (at_stats[at_stats["AwayTeam"] == at].sum().FTAG + at_stats[at_stats["HomeTeam"] == at].sum().FTHG)
    epl_stat.at[i, 'pastHxG'] = (ht_stats[ht_stats["AwayTeam"] == ht].sum().AxG + ht_stats[ht_stats["HomeTeam"] == ht].sum().HxG)
    epl_stat.at[i, 'pastAxG'] = (at_stats[at_stats["AwayTeam"] == at].sum().AxG + at_stats[at_stats["HomeTeam"] == at].sum().HxG)
    epl_stat.at[i, 'pastHS'] = (ht_stats[ht_stats["AwayTeam"] == ht].sum().AS + ht_stats[ht_stats["HomeTeam"] == ht].sum().HS)
    epl_stat.at[i, 'pastAS'] = (at_stats[at_stats["AwayTeam"] == at].sum().AS + at_stats[at_stats["HomeTeam"] == at].sum().HS)
    epl_stat.at[i, 'pastHST'] = (ht_stats[ht_stats["AwayTeam"] == ht].sum().AST + ht_stats[ht_stats["HomeTeam"] == ht].sum().HST)
    epl_stat.at[i, 'pastAST'] = (at_stats[at_stats["AwayTeam"] == at].sum().AST + at_stats[at_stats["HomeTeam"] == at].sum().HST)
```

```python
def HomePoint(row):
    if(row.Result == 'H'):
        return 3
    elif(row.Result == 'A'):
        return 0
    elif(row.Result == 'D'):
        return 1
    else:
        return 0
```

```python
epl_stat["HomePoint"] = epl_stat.apply(lambda row: HomePoint(row),axis=1)
```

```python
epl_stat
```

```python
def AwayPoint(row):
    if(row.Result == 'H'):
        return 0
    elif(row.Result == 'A'):
        return 3
    elif(row.Result == 'D'):
        return 1
    else:
        return 0
```

```python
epl_stat["AwayPoint"] = epl_stat.apply(lambda row: AwayPoint(row),axis=1)
```
