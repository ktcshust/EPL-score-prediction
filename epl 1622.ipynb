```python
import requests
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv("2016-2017.csv")
df2 = pd.read_csv("2017-2018.csv")
df3 = pd.read_csv("2018-2019.csv")
df4 = pd.read_csv("2019-2020.csv")
df5 = pd.read_csv("2020-2021.csv")
df6 = pd.read_csv("2021-2022.csv")
```

```python
df1
```

```python
df = pd.concat([df1, df2, df3, df4, df5, df6])
```

```python
epl_stat = df.reset_index()
epl_stat
```

```python
def HomePoint(row):
    if(row.FTR == 'H'):
        return 3
    elif(row.FTR == 'A'):
        return 0
    elif(row.FTR == 'D'):
        return 1
    else:
        return 0
```

```python
epl_stat["HomePoint"] = epl_stat.apply(lambda row: HomePoint(row),axis=1)
```

```python
def AwayPoint(row):
    if(row.FTR == 'A'):
        return 3
    elif(row.FTR == 'H'):
        return 0
    elif(row.FTR == 'D'):
        return 1
    else:
        return 0
epl_stat["AwayPoint"] = epl_stat.apply(lambda row: AwayPoint(row),axis=1)
```

```python
epl_stat["pastHG"] = 0.0
epl_stat["pastAG"] = 0.0
epl_stat["pastHGC"] = 0.0
epl_stat["pastAGC"] = 0.0
epl_stat['pastHP'] = 0.0
epl_stat['pastAP'] = 0.0
```

```python
k = 2
for i in range(0,2280):
    row = epl_stat.loc[i]
    ht = row.HomeTeam
    at = row.AwayTeam
    ht_stats = epl_stat.loc[0:i-1][(epl_stat.HomeTeam == ht) | (epl_stat.AwayTeam == ht)].tail(k)
    at_stats = epl_stat.loc[0:i-1][(epl_stat.HomeTeam == at) | (epl_stat.AwayTeam == at)].tail(k)
    epl_stat.at[i, 'pastHG'] = (ht_stats[ht_stats["AwayTeam"] == ht].sum().FTAG + ht_stats[ht_stats["HomeTeam"] == ht].sum().FTHG)
    epl_stat.at[i, 'pastAG'] = (at_stats[at_stats["AwayTeam"] == at].sum().FTAG + at_stats[at_stats["HomeTeam"] == at].sum().FTHG)
    epl_stat.at[i, 'pastHGC'] = (ht_stats[ht_stats["AwayTeam"] == ht].sum().FTHG + ht_stats[ht_stats["HomeTeam"] == ht].sum().FTAG)
    epl_stat.at[i, 'pastAGC'] = (at_stats[at_stats["AwayTeam"] == at].sum().FTHG + at_stats[at_stats["HomeTeam"] == at].sum().FTAG)
    epl_stat.at[i, 'pastHP'] = (at_stats[at_stats["AwayTeam"] == at].sum().AwayPoint + at_stats[at_stats["HomeTeam"] == at].sum().HomePoint)
    epl_stat.at[i, 'pastAP'] = (at_stats[at_stats["AwayTeam"] == at].sum().AwayPoint + at_stats[at_stats["HomeTeam"] == at].sum().HomePoint)
```

```python
epl_stat['HomeFullPoint'] = 0.0
epl_stat['AwayFullPoint'] = 0.0
```

```python
for t in range(0,6):
    for i in range(380*t,380*t+380):
        row = epl_stat.loc[i]
        ht = row.HomeTeam
        at = row.AwayTeam
        ht_stats = epl_stat.loc[380*t:i-1][(epl_stat.HomeTeam == ht) | (epl_stat.AwayTeam == ht)]
        at_stats = epl_stat.loc[380*t:i-1][(epl_stat.HomeTeam == at) | (epl_stat.AwayTeam == at)]
        epl_stat.at[i, 'HomeFullPoint'] = (ht_stats[ht_stats["AwayTeam"] == ht].sum().AwayPoint + ht_stats[ht_stats["HomeTeam"] == ht].sum().HomePoint)
        epl_stat.at[i, 'AwayFullPoint'] = (at_stats[at_stats["AwayTeam"] == at].sum().AwayPoint + at_stats[at_stats["HomeTeam"] == at].sum().HomePoint)
            
        
```

```python
epl_stat['HomeGoal'] = 0.0
epl_stat['AwayGoal'] = 0.0
epl_stat['HomeConceded'] = 0.0
epl_stat['AwayConceded'] = 0.0
```

```python
for t in range(0,6):
    for i in range(380*t,380*t+380):
        row = epl_stat.loc[i]
        ht = row.HomeTeam
        at = row.AwayTeam
        ht_stats = epl_stat.loc[380*t:i-1][(epl_stat.HomeTeam == ht) | (epl_stat.AwayTeam == ht)]
        at_stats = epl_stat.loc[380*t:i-1][(epl_stat.HomeTeam == at) | (epl_stat.AwayTeam == at)]
        epl_stat.at[i, 'HomeGoal'] = (ht_stats[ht_stats["AwayTeam"] == ht].sum().FTAG + ht_stats[ht_stats["HomeTeam"] == ht].sum().FTHG)
        epl_stat.at[i, 'AwayGoal'] = (at_stats[at_stats["AwayTeam"] == at].sum().FTAG + at_stats[at_stats["HomeTeam"] == at].sum().FTHG)
        epl_stat.at[i, 'HomeConceded'] = (ht_stats[ht_stats["AwayTeam"] == ht].sum().FTHG + ht_stats[ht_stats["HomeTeam"] == ht].sum().FTAG)
        epl_stat.at[i, 'AwayConceded'] = (at_stats[at_stats["AwayTeam"] == at].sum().FTHG + at_stats[at_stats["HomeTeam"] == at].sum().FTAG)
            
```

```python
epl_stat['HomeAvgPoint'] = 0.0
epl_stat['AwayAvgPoint'] = 0.0
for i in range(0,2280):
    row = epl_stat.loc[i]
    ht = row.HomeTeam
    at = row.AwayTeam
    ht_stats = epl_stat.loc[0:i-1][(epl_stat.HomeTeam == ht) | (epl_stat.AwayTeam == ht)]
    at_stats = epl_stat.loc[0:i-1][(epl_stat.HomeTeam == at) | (epl_stat.AwayTeam == at)]
    epl_stat.at[i, 'HomeAvgPoint'] = (ht_stats[ht_stats["AwayTeam"] == ht].sum().AwayPoint + ht_stats[ht_stats["HomeTeam"] == ht].sum().HomePoint)/len(ht_stats)
    epl_stat.at[i, 'AwayAvgPoint'] = (at_stats[at_stats["AwayTeam"] == at].sum().AwayPoint + at_stats[at_stats["HomeTeam"] == at].sum().HomePoint)/len(at_stats)
```

```python
epl_stat = epl_stat.replace(np.nan, 0)
```

```python
data = epl_stat.reset_index()
data
```
