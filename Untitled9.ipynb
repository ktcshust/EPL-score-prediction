```python
epl = w[['level_0','Wk','Home','Away','FTHG','FTAG','xG','xG.1']]
epl = epl.drop([60, 61, 62, 63, 64, 65, 66, 67, 68, 69, 74, 77, 79, 114]).reset_index()
```

```python
epl["FTHG"] = epl['FTHG'].apply(int)
epl["FTAG"] = epl['FTAG'].apply(int)
epl['xG'] = epl['xG'].replace(np.nan, -1)
epl['xG.1'] = epl['xG.1'].replace(np.nan, -1)
epl['xG'] = epl['xG'].apply(float)
epl['xG.1'] = epl['xG.1'].apply(float)
epl['xG'] = epl['xG'].replace(-1, 0)
epl['xG.1'] = epl['xG.1'].replace(-1, 0)
```

```python
epl['HomeTeam'] = epl['Home']
epl['AwayTeam'] = epl['Away']
feat_table = epl[['level_0','Wk','HomeTeam','AwayTeam','FTHG','FTAG','xG','xG.1']]
feat_table
```

```python
def Result(row):
    if(row.FTHG > row.FTAG):
        return 'H'
    elif(row.FTHG < row.FTAG):
        return 'A'
    elif(row.FTHG == -1):
        return np.nan
    else:
        return 'D'
feat_table["Result"] = feat_table.apply(lambda row: Result(row),axis=1)
feat_table
```

```python
epl_stat["pastHG"] = 0.0
epl_stat["pastAG"] = 0.0
epl_stat["pastHxG"] = 0.0
epl_stat["pastAxG"] = 0.0
```

```python
k = 3
for i in range(0,366):
    row = epl_stat.loc[i]
    ht = row.HomeTeam
    at = row.AwayTeam
    ht_stats = epl_stat.loc[0:i-1][(epl_stat.HomeTeam == ht) | (epl_stat.AwayTeam == ht)].tail(k)
    at_stats = epl_stat.loc[0:i-1][(epl_stat.HomeTeam == at) | (epl_stat.AwayTeam == at)].tail(k)
    epl_stat.at[i, 'pastHG'] = (ht_stats[ht_stats["AwayTeam"] == ht].sum().FTAG + ht_stats[ht_stats["HomeTeam"] == ht].sum().FTHG)
    epl_stat.at[i, 'pastAG'] = (at_stats[at_stats["AwayTeam"] == at].sum().FTAG + at_stats[at_stats["HomeTeam"] == at].sum().FTHG)
    epl_stat.at[i, 'pastHxG'] = (ht_stats[ht_stats["AwayTeam"] == ht].sum().AxG + ht_stats[ht_stats["HomeTeam"] == ht].sum().HxG)
    epl_stat.at[i, 'pastAxG'] = (at_stats[at_stats["AwayTeam"] == at].sum().AxG + at_stats[at_stats["HomeTeam"] == at].sum().HxG)
```
